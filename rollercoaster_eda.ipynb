#### Step 2: Data Preparation

This step involves,
- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Identifying and handling missing values
- Renaming columns
- Feature creation

```python
df.columns
```

```python
# dropping irrelevant columns
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', #'Opening date','Type', 
    'Manufacturer', 
#     'Height restriction', 'Model', 'Height','Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
#      'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
#      'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
#      'Track layout', 'Fastrack available', 'Soft opening date.1','Closing date','Opened', 
#     'Replaced by', 'Website',
#        'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
#        'Single rider line available', 'Restraint Style',
#        'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude',
    'Type_Main',
       'opening_date_clean', 
#     'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       
    'speed_mph', #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.columns
```

```python
# converting the data types

df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
df.dtypes

col = ['coaster_name', 'Location', 'Status', 'Manufacturer', 'Type_Main']

for col in col:
    df[col] = df[col].astype('string')
```

```python
df.info()
```

```python
# renaming the columns
df = df.rename(columns = {'coaster_name':'roller_coaster_name','Location' :'location', 'Status' : 'status', 'Manufacturer' : 'manufacturer',
        'Type_Main' : 'material_used', 'opening_date_clean' : 'opening_date',
    'Inversions_clean' : 'inversion', 'Gforce_clean':'gforce'})
df
```

```python
# look for missing values and their frequency
df.isna().sum()
```

```python
#look for duplicate columns
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset="roller_coaster_name")].head(5)
```

```python
#checking examples of duplicate
df.query('roller_coaster_name == "Crystal Beach Cyclone"')
```

```python
df = df.loc[~df.duplicated(subset=["roller_coaster_name", "location", "opening_date"])].reset_index(drop=True).copy()
df
```

#### Step 4: Feature Relationship

This step involves understanding of relationship between more than two variables using both Pandas visualisation libraries (matplotlib and seaborn)
- Scatter Plot
- Heatmap Correlation

- Groupby comparison


```python
# comparing two variables against each other using matplotlib
df.plot(kind='scatter',
       x='speed_mph',
       y='height_ft',
       title='coaster speed vs height');
```

```python
# compare multiple features in the dataset using the pairplot
sns.pairplot(df, 
             vars=['year_introduced', 'speed_mph', 'height_ft', 'inversion','gforce'],
             hue = 'material_used'
            );
```

```python
#correlation
df_corr = df[['year_introduced', 'speed_mph', 'height_ft', 'inversion','gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot = True);
```

#### Step 5 : Ask a Question about the data

What are the locations with fastest roller coasters with minimum of 10 coasters on that location?

```python
ax = df.query('location != "Other"')\
    .groupby('location')['speed_mph']\
    .agg(['mean', 'count'])\
    .query('count >= 10')\
    .sort_values('mean')['mean']\
    .plot( kind = 'bar', 
          title = 'Average coaster speed by location'
        );
ax.set_xlabel('average coaster speed')
```
